```python
import csv

def most_common(L):
    if len(L) > 0:
        return max(set(L), key=L.count)
    else: 
        return 0

# information on the animal services intake for Los Angeles
f = open('Animal_Services_Intake_Data.csv')
csv_f = csv.reader(f)

numCats = 0
numDogs = 0
numHorses = 0
numOther = 0
numTotalPets = sum(1 for row in csv_f) - 1
catBreeds = []
catWhere = []

f.seek(0)

for row in csv_f:
    #print row[5]
    if row[5] == 'CAT':
        numCats += 1
        catBreeds.append(row[7])
        catWhere.append(row[0])
    elif row[5] == 'DOG':
        numDogs += 1
    elif row[5] == 'EQUINE':
        numHorses += 1
    else:
        numOther += 1
        
outputFile = open("AnimalIntakeStats.txt", "w")
outputFile.write("numTotalPets = " + str(numTotalPets) + ';\n')
outputFile.write("numCats = " + str(numCats) + ";\n")
outputFile.write("numDogs = " + str(numDogs) + ";\n")
outputFile.write("numHorses = " + str(numHorses) + ";\n")
outputFile.write("numOther = " + str(numOther) + ";\n")
#outputFile.write('catBreeds = ["' + '","'.join(catBreeds) + '"];\n')
#outputFile.write('catWhere = ["'+ '","'.join(catWhere) + '"];\n')
outputFile.write("domestic shorthair: " + str(catBreeds.count("DOMESTIC SH")) + "\n")
outputFile.write("domestic longhair: " + str(catBreeds.count("DOMESTIC LH")) + "\n")
outputFile.write("bombay: " + str(catBreeds.count("BOMBAY")) + "\n")
outputFile.write("persian: " + str(catBreeds.count("PERSIAN")) + "\n")
outputFile.write("maine coon: " + str(catBreeds.count("MAINE COON")) + "\n")
outputFile.write("siamese: " + str(catBreeds.count("SIAMESE")) + "\n")
outputFile.write("amer sh: " + str(catBreeds.count("AMER SH")) + "\n")
outputFile.write("russian blue: " + str(catBreeds.count("RUSSIAN BLUE")) + "\n")
outputFile.write("abyssinian: " + str(catBreeds.count("ABYSSINIAN")) + "\n")
outputFile.write("ragdoll: " + str(catBreeds.count("RAGDOLL")) + "\n")


f.close()
outputFile.close()
    
print 'Number of total pets: '
print numTotalPets
print 'Number of cats surrendered: '
print numCats
print 'Most common cat breed surrendered: '
print most_common(catBreeds)
print 'Most common location in LA of cat breed surrendered: '
print most_common(catWhere)
print 'Breeds of cats surrendered: '
print catBreeds
print 'Locations of cats surrendered: '
print catWhere
```
